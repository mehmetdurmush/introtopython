```python
grades={}
i=0
for i in range(0,6):
    name=input("Enter your name: ")
    midterm = float(input("Enter your midterm grade: "))
    project = float(input("Enter your project grade: "))
    final= float(input("Enter your final grade: "))
    passingGrade=midterm*0.3+project*0.3+final*0.4
    print(passingGrade)
    grades[name] = passingGrade
    i += 1
    if i==5:
        break

sort_grades = sorted(grades.items(), key=lambda x: x[1], reverse=True)
print("\n Grades")
for i in sort_grades:
    print(i[0],i[1])
```
